```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#  3D-
fig = plt.figure(figsize=(14, 6))

# ===========================================
#  1:     
# ===========================================
ax1 = fig.add_subplot(121, projection='3d')

#   O
O = np.array([0, 0, 0])

#     O
t = np.linspace(-2, 2, 100)

#  1:   X
ax1.plot(t, 0*t, 0*t, 'r-', linewidth=2, label=': x-')

#  2:   Y
ax1.plot(0*t, t, 0*t, 'g-', linewidth=2, label=': y-')

#  3:   x=y=z
ax1.plot(t, t, t, 'b-', linewidth=2, label=': x=y=z')

#     O ( )
ax1.plot(t, -t, 0.5*t, 'm-', linewidth=2, label=':  O')

#  O
ax1.scatter([0], [0], [0], color='black', s=100, label=' O')

ax1.set_title(' 1:      O')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.legend()
ax1.view_init(elev=20, azim=45)

# ===========================================
#  2:     
# ===========================================
ax2 = fig.add_subplot(122, projection='3d')

#  z=0 ()
u, v = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10))
z_plane = 0*u
ax2.plot_surface(u, v, z_plane, alpha=0.3, color='yellow', label=' ')

#    
#  1: y=0   z=0
ax2.plot(t, 0*t, 0*t, 'r-', linewidth=2, label=': y=0')

#  2: x=0   z=0
ax2.plot(0*t, t, 0*t, 'g-', linewidth=2, label=': x=0')

#  3: y=x   z=0
ax2.plot(t, t, 0*t, 'b-', linewidth=2, label=': y=x')

#  4: y=-x   z=0
ax2.plot(t, -t, 0*t, 'm-', linewidth=2, label=': y=-x')

ax2.set_title(' 2:     ')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.legend()
ax2.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()

# :   "" -     O
fig2 = plt.figure(figsize=(8, 6))
ax3 = fig2.add_subplot(111, projection='3d')

#    O
ax3.plot(t, 0*t, 0*t, 'r-', linewidth=2, label='')
ax3.plot(0*t, t, 0*t, 'g-', linewidth=2, label='')
ax3.plot(t, t, t, 'b-', linewidth=2, label='')

#  O
ax3.scatter([0], [0], [0], color='black', s=100, label=' O')

# ,    O -     ,   !
#    (1,0,0)   (0,1,0) -    XY
ax3.plot([1, 0], [0, 1], [0, 0], 'm--', linewidth=3, label=' (  O)')

ax3.set_title(':     O\n    ,   !')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Z')
ax3.legend()
ax3.view_init(elev=20, azim=45)

plt.tight_layout()
plt.show()
```
